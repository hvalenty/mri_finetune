```python
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

Functions for plotting different numbers of slices

```python
def plot_first_ten(path):
    series = np.load(path)

    fig = plt.figure(figsize=(20, 6))

    for i, image in enumerate(series):
        ax = fig.add_subplot(1, 10, i+1, xticks=[], yticks=[])
        plt.imshow(image, cmap='gray')
        plt.axis('off')

def plot_first_thirty(path):
    series = np.load(path)

    fig = plt.figure(figsize=(20, 6))

    for i, image in enumerate(series):
        ax = fig.add_subplot(3, 10, i+1, xticks=[], yticks=[])
        plt.imshow(image, cmap='gray')
        plt.axis('off')

def plot_first_onefifty(path):
    series = np.load(path)

    fig = plt.figure(figsize=(20, 6))

    for i, image in enumerate(series):
        ax = fig.add_subplot(10, 15, i+1, xticks=[], yticks=[])
        plt.imshow(image, cmap='gray')
        plt.axis('off')
```
